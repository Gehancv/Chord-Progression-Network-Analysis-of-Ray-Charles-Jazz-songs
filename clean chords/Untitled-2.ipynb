```python
import pandas as pd
```

```python
#RAY CHARLES

#Dictionary
df2 = pd.read_csv('node.csv')
df = pd.read_csv('edge.csv')
mapping_dict = dict(zip(df2['Id'], df2['Label']))
#source and target
df['Source'] = df['Source'].replace(mapping_dict)
df['Target'] = df['Target'].replace(mapping_dict)
#Cleaning equivalent chords:
df3=pd.read_csv('clean_chords.csv')
#mapping and changing values
mapping_dict2 = dict(zip(df3['from_l'], df3['to_l']))
df['Source'] = df['Source'].replace(mapping_dict2)
df['Target'] = df['Target'].replace(mapping_dict2)
#removing self loops
df = df[df['Source'] != df['Target']]
df
```

```python
df.to_csv('RCnew_edges.csv',index=False)
```
