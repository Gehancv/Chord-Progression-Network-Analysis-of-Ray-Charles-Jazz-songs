```python
import pandas as pd
import os
import csv
import numpy as np
```

```python

input_folder_path = 'C:\\Users\\MS-XUserPC\\Downloads\\Songs'
output_folder_path = 'C:\\Users\\MS-XUserPC\\Downloads\\output'
```

```python
# Iterate through each file in the input folder
for filename in os.listdir(output_folder_path):
    if filename.endswith(".csv"):
        input_file_path = os.path.join(input_folder_path, filename)

        # Construct the output CSV file path based on the input file name
        output_csv_filename = os.path.splitext(filename)[0] + '.csv'
        output_csv_path = os.path.join(output_folder_path, output_csv_filename)

        # List to store lines and set to store unique pairs
        lines = []
        unique_pairs = set()

        # Open each text file for reading
        with open(input_file_path, 'r') as txt_file:
            for _ in range(4):
                next(txt_file)  # eliminate first four lines

            # Read the rest of the lines from the text file
            for line in txt_file:
                columns = line.strip().split('\t')
                if len(columns) == 2:
                    pair = (columns[0], columns[1])
                    if pair not in unique_pairs:
                        unique_pairs.add(pair)
                        lines.append(line)

  # Open the CSV file for writing
        with open(output_csv_path, 'w', newline='') as csv_file:
            csv_writer = csv.writer(csv_file, delimiter=',')

            # Write the lines to the CSV file
            csv_writer.writerows(line.strip().split('\t') for line in lines)
```
